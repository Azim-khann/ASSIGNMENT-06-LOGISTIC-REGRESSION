#### LIBRARIES

```python
import pandas as pd
import numpy as np
```

#### LOAD/READ THE DATASET

```python
bank_full = pd.read_excel('bank_full.xlsx')
bank_full
```

#### TAKE BASIC INFORMATION FROM THE DATA & EDA

```python
bank_full.info()
```

```python
bank_full.describe()
```

#### One-Hot Encoding of categrical variables

```python
data = pd.get_dummies(bank_full,columns=['job','marital','education','contact','poutcome'])
data 
```

```python
pd.set_option("display.max.columns", None)
data
```

```python
data.info()
```

#### Custom Binary Encoding of Binary o/p variables 

```python
data['default'] = np.where(data['default'].str.contains("yes"), 1, 0)
data['housing'] = np.where(data['housing'].str.contains("yes"), 1, 0)
data['loan'] = np.where(data['loan'].str.contains("yes"), 1, 0)
data['y'] = np.where(data['y'].str.contains("yes"), 1, 0)
data
```

#### Find and Replace Encoding for month categorical varaible

```python
data['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
data = data.replace(order)
```

```python
data
```

#### Model Building

```python
x = pd.concat([data .iloc[:,0:11],data.iloc[:,12:]],axis=1)
y = data.iloc[:,11]
```

#### Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x,y)
```

#### Model Predictions

```python
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

#### Testing Model Accuracy

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

#### The model accuracy is calculated by (a+d)/(a+b+c+d)

```python
(39107+1282)/(39107+815+4007+1282)
```

```python
classifier.predict_proba(x)[:,1] 
```

#### ROC Curve plotting and finding AUC value

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)
```

```python
plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()
print('auc accuracy:',auc)
```
